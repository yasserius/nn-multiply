```python
import glob as glob

DIRS = glob.glob("../runs/*")
DIRS.remove('../runs\\README.md')
DIRS
```

```python
import os
import glob
import json
import numpy as np

def load_json(f):
    with open(f, 'r') as json_f:
        return json.loads(json_f.read())

def get_json_logs(base_dir):
    files = []
    for f in glob.glob(base_dir + f'/*/*.json'):
        print(f)
        files.append(load_json(f))
    return files
```

```python
import plotly.graph_objects as go
from scipy import signal

f2_all = get_json_logs(DIRS[2])
f2_all += get_json_logs(DIRS[1])
f2_all += get_json_logs(DIRS[3])


fig = go.Figure()

for f in f2_all:
    model_name = f['config_name'].split('-')[-1]

    line_color = "rgba(0,0,0,0.2)"
    if model_name == '50000_nodes_lr_0_001': line_color = "rgba(255,4,4,1)"
    fig.add_trace(
      go.Scatter(y=signal.savgol_filter(f['r2'], 51, 4),
                mode='lines',
                name=model_name,
                hovertemplate =
                  'R^2: %{y:.2f}<br>' +
                  'Epoch: %{x}<br>' +
                  'Highest mutiplicant: 100',
                line=dict(
                  color=line_color,   
                ),
                showlegend=False)
    )

fig.update_layout(
    title="Performance of single layer FCNs on reversed logarithm and addition task",
    xaxis_title="Epochs",
    yaxis_title="R^2",
    # legend_title="Model",
    yaxis_range=[0.0, 0.6],
    template="ggplot2"
)

fig.show()
fig.write_html("r2_plot_single_layer.html")
```

```python
f3_all = get_json_logs(DIRS[3])
f3_all += get_json_logs(DIRS[4])

fig = go.Figure()

for f in f3_all:
    model_name = f['config_name'].split('-')[-1]

    line_color = "rgba(0,0,0,0.2)"
    if model_name == '500_500_nodes_0_0001': line_color = "rgba(255,4,4,1)"
    fig.add_trace(
      go.Scatter(y=signal.savgol_filter(f['r2'], 51, 4),
                mode='lines',
                name=model_name,
                hovertemplate =
                  'R^2: %{y:.2f}<br>' +
                  'Epoch: %{x}<br>' +
                  'Highest mutiplicant: 100',
                line=dict(
                  color=line_color,   
                ),
                showlegend=False)
    )

fig.update_layout(
    title="Performance of two layered FCNs on reversed logarithm and addition task",
    xaxis_title="Epochs",
    yaxis_title="$R^2$",
    # legend_title="Model",
    yaxis_range=[0.0, 0.6],
    template="ggplot2"
)

fig.show()
fig.write_html("r2_plot_single_layer.html")
```
